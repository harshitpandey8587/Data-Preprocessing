```python
d_Question = []
d_Answer = []

for conversation in df["conversations"]:
    for i in conversation["MessageList"]:
        if i["from"] != "8:live:.cid.4933a22bf564744":
            data = {'conversationid': i["from"], 'displayName': i["displayName"], 'originalarrivaltime': i["originalarrivaltime"], 'content':
                i["content"]}
            d_Question.append(data)
        else:
            data_ans = {'conversationid': i["from"], 'displayName': i["displayName"], 'originalarrivaltime': i["originalarrivaltime"], 'content':
                i["content"]}
            d_Answer.append(data_ans)
```

```python
df_question = pd.DataFrame(d_Question)
df_answer = pd.DataFrame(d_Answer)
```

# Converting samle data and intent into the NeMo format

```python
import pandas as pd
```

```python
df_intents = pd.read_excel("Intents_sample_data.xlsx", sheet_name="unique_intents")
```

```python
df_intents
```

```python
intent_dict = df_intents.to_dict()["Intent"]
```

```python
intent_dict
```

```python
new_intent_dict = dict([(value, key) for key, value in intent_dict.items()]) 
```

```python
new_intent_dict
```

```python
df_sample = pd.read_csv("sample_data_preprocessing.csv")
```

```python
df_sample
```

```python
df_sample.drop(["id","displayName","originalarrivaltime"], inplace=True, axis=1)
```

```python
df_sample
```

```python
df_sample.drop("slot", inplace=True, axis=1)
```

```python
df_sample
```

```python
df_sample['Intent_number'] = df_sample['intent'].map(new_intent_dict)
df_sample
```

```python
df_sample.info()
```

```python
df_sample["Intent_number"] = df_sample["Intent_number"].astype(int)
```

```python
df_sample
```

```python
import re
content_updated = []
for content in df_sample["content"]:
    print("before   ", content)
    result = ' '.join(content.split())
    print('-----------------')
    print(result)
    result = re.sub(r"[\t\n\r]*", "", content)
    result = re.sub(r"[\n]*", "", result)
    content_updated.append(result)
    print("after   ", content)
```

```python
df_sample["content_updated"] = content_updated
```

```python
df_sample
```

```python
df_intents.to_csv("dict.intents.csv", index=False)
```

```python
df_sample.drop(["intent","content"], inplace=True, axis=1)
```

```python
df_sample = df_sample[["content_updated","Intent_number"]]
```
